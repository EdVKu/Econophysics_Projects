# Demostración Computacional de Ley de Winger de Semi-Circulo:

La Ley de Winger de Semicírculo postula que, dado un ensamble de matrices aleatorias $\{A_i\}_{i = 0}^{m},\ A_i \in M_{n\times n}(F)$ sobre un campo, entonces su distribución de Eigenvalores $\rho(E)$ se aproxima a la siguiente función:  

$$\rho(E) =\frac{2\sqrt{R²+E²}}{\pi R²}$$

Para una n suficientemente grande.

Se llevará a cabo la demostración computacional con $n \in \{5,20,30,100\}$

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1300)
x = []
def goe(n):
    x = np.random.uniform(0,1,(n,n))
    x = np.sqrt(0.5)*(x+np.transpose(x))
    return x
def goe2(n):
    x = np.random.exponential(1,(n,n))
    x = np.sqrt(0.5)*(x+np.transpose(x))
    return x
```

```python

# Generar un conjunto de m GOEs de dimensión n
def multgoe(a,b):
    p = []
    for i in range(0,a):
        p.append(goe(b))
    return p
def multgoe2(a,b):
    p = []
    for i in range(0,a):
        p.append(goe2(b))
    return p
#Obtener eigenvalores y guardar los mxn eigenvalores en un sólo vector columna



n5 =multgoe(50, 5)
n20 = multgoe(50, 20)
n30 = multgoe(50, 30)
n100 = multgoe(50, 115)
m5 =multgoe2(50, 5)
m20 = multgoe2(50, 20)
m30 = multgoe2(50, 30)
m100 = multgoe2(50, 115)

def eig(m):
    p = []
    for i in range(0, np.shape(m)[0]):
        diago = np.linalg.eig(m)[0][i]
        media = np.mean(diago)
        desvest = np.std(diago)
        for j in range(0, np.shape(m)[1]):
            p.append((float(diago[j])-media)/desvest)
    return np.transpose(p)



#Comparar eigenvalores con ley de Winger de semicírculo
```

```python
Eig = eig(n5)
R = np.min(Eig)
print(R)
xc = np.linspace(-R, R, num=1000)
y = 2*np.sqrt(R**2-(xc)**2)/(np.pi*R**2)
plt.hist(Eig,bins=50, density = True)
plt.plot(xc,y)
```

```python
Eig = eig(n20)
R = np.min(Eig)
xc = np.linspace(-R, R, num=1000)
y = 2*np.sqrt(R**2-(xc)**2)/(np.pi*R**2)

plt.hist(Eig,bins=50, density = True)
plt.plot(xc,y)
```

```python
Eig = eig(n30)
R = np.min(Eig)
xc = np.linspace(-R, R, num=1000)
y = 2*np.sqrt(R**2-(xc)**2)/(np.pi*R**2)
plt.hist(Eig,bins=50, density = True)
plt.plot(xc,y)
```

```python
Eig = eig(n100)
R = np.min(Eig)
xc = np.linspace(-R, R, num=1000)
y = 2*np.sqrt(R**2-(xc)**2)/(np.pi*R**2)
plt.hist(Eig,bins=50, density = True)
plt.plot(xc,y)
```

Notamos que para la distribución uniforme se conserva de manera directa la relación de semicírculo de Winger.

De igual manera, para la distribución exponencial, sus eigenvalores se distribuyen aproximándose a la distribución de Winger.

Para ambas distribuciones, este ajuste fue más notorio una vez llegamos a los valores más altos. De igual manera, para ambas distribuciones, se cumple que existe una cantidad no nula de valores que no logran ser ajustados (_outliers_).

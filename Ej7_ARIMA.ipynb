```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pmdarima.arima import auto_arima
from math import *
#arima-x
```

```python
ticks = pd.read_csv('/content/drive/MyDrive/Classroom/TSFMT-II(2024-1) Física/ParisIndex-2016-2023.csv', index_col=0)
ticks2 = ticks.to_numpy()[0,1:]
plt.plot(ticks2, color = 'red')
```

```python
tickslog = [log(ticks2[i]) for i in range(len(ticks2))]
```

```python
rets = [(ticks2[i+1]-ticks2[i])/ticks2[i] for i in range(len(ticks2)-1)]

plt.plot(rets)
```

```python
train = ticks2[:1587]
test = ticks2[1587:]
train2 = rets[:int(0.8*len(rets))]
test2 = rets[int(0.8*len(rets)):]
train3 = tickslog[:int(0.8*len(tickslog))]
test3 = tickslog[int(0.8*len(tickslog)):]
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(test)
ax[0,1].plot(train)
ax[1,0].plot(test3)
ax[1,1].plot(train3)
```

```python
modelo_arima = auto_arima(train, seasonal=False, stepwise=False, approximation= False)
modelo_arimaRe = auto_arima(train2, seasonal=False, stepwise=False, approximation= False)

modelo_arimaLog = auto_arima(train3, seasonal=False, stepwise=False, approximation= False)
```

```python
test = pd.DataFrame(test)
test2 = pd.DataFrame(test2)
test3 = pd.DataFrame(test3)
```

```python
pred = pd.DataFrame(modelo_arima.predict(n_periods = 401))

pred2 = pd.DataFrame(modelo_arimaRe.predict(n_periods=400))

pred3 = pd.DataFrame(modelo_arimaLog.predict(n_periods = 401))
```

```python
plt.plot(pred)
plt.plot(test)
plt.title("Serie temporal para predicción de conjunto de prueba")
plt.show()
```

```python
plt.plot(pred3)
plt.plot(test3)
plt.title("Serie temporal para predicción de log(conjunto de prueba)")
plt.show()
```

A partir de aquí, se graficaron las funciones de autocorrelación (ACF) y las funciones de autocorrelación parcial (PACF) para la predicción y los conjuntos de prueba (remanentes, serie temporal y log(serie temporal)).
